## Profiling Serial `lennard_jones_potential.py`

```
$ python -m cProfile -o output_serial.prof lennard_jones_potential.py 
Serial execution on 2000 points
Runtime: 20.819 seconds

V_total = 23630468668.9
```

````python
import pstats
p = pstats.Stats('output_serial.prof')
p.print_stats()
````

````python
p.strip_dirs()
p.sort_stats('ncalls')
p.print_stats(20)
````

````python
p.strip_dirs()
p.sort_stats('cumtime')
p.print_stats(10)
````

````python
p.strip_dirs()
p.sort_stats('tottime')
p.print_stats(10)
````

````python
p.print_stats("lennard_jones_potential.py:")
````
